```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os

# Set image path
img_path = "/kaggle/input/datase/Test/actinic keratosis/ISIC_0024470_input.jpg"

# Derive true label from folder name
true_label = os.path.basename(os.path.dirname(img_path)).lower()

# Preprocess image
img = Image.open(img_path).convert('L')     # Grayscale
img = img.resize((288, 225))                # Resize to match training size
img_array = np.array(img)
img_array = np.expand_dims(img_array, axis=-1)  # (225, 288, 1)
img_array = np.expand_dims(img_array, axis=0)   # (1, 225, 288, 1)
img_array = img_array / 255.0                   # Normalize

# Class labels
class_labels = [
    "actinic keratosis", "basal cell carcinoma", "dermatofibroma",
    "melanoma", "nevus", "pigmented benign keratosis",
    "seborrheic keratosis", "squamous cell carcinoma", "vascular lesion"
]

# Predict
predictions = model.predict(img_array)
predicted_index = np.argmax(predictions[0])
predicted_label = class_labels[predicted_index]
confidence = predictions[0][predicted_index]

# Display result
plt.imshow(img, cmap='gray')
plt.title(f"True: {true_label}\nPredicted: {predicted_label} ({confidence:.2f})", fontsize=12)
plt.axis('off')
plt.show()

print("Image shape:", img_array.shape)
print("Min, Max pixel values:", img_array.min(), img_array.max())
print("Predicting...")
predictions = model.predict(img_array)
print("Predictions:", predictions)
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# -----------------------------
# Load Original Image
# -----------------------------
img_path = "/kaggle/input/augmented-skincancer-isic/skin_cancer/skin_cancer/Test/actinic keratosis/ISIC_0010512.jpg"
true_label = os.path.basename(os.path.dirname(img_path)).lower()

# Read original using PIL & keep it RGB
original_img = Image.open(img_path).convert("RGB")
gray_img = original_img.convert('L')

# -----------------------------
# Enhancement using CLAHE
# -----------------------------
gray_array = np.array(gray_img)
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
enhanced_array = clahe.apply(gray_array)

enhanced_img = Image.fromarray(enhanced_array)
enhanced_img = enhanced_img.resize((288, 225))     # Resize to model input size

# -----------------------------
# Data Augmentation
# -----------------------------
aug = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True
)

# Prepare augmented samples
aug_input = np.expand_dims(np.array(enhanced_img), axis=(0, -1)) / 255.0   # (1,225,288,1)
aug_generator = aug.flow(aug_input, batch_size=1)

# Generate 3 augmented samples for visualization
augmented_samples = [next(aug_generator)[0].reshape(225, 288) for _ in range(3)]

# -----------------------------
# Preprocess for Prediction (Enhanced Image)
# -----------------------------
img_array = np.array(enhanced_img).astype("float32") / 255.0
img_array = np.expand_dims(img_array, axis=-1)     # (225,288,1)
img_array = np.expand_dims(img_array, axis=0)      # (1,225,288,1)

# -----------------------------
# Class Labels
# -----------------------------
class_labels = [
    "actinic keratosis", "basal cell carcinoma", "dermatofibroma",
    "melanoma", "nevus", "pigmented benign keratosis",
    "seborrheic keratosis", "squamous cell carcinoma", "vascular lesion"
]

# -----------------------------
# Predict
# -----------------------------
predictions = model.predict(img_array)
predicted_index = np.argmax(predictions[0])
predicted_label = class_labels[predicted_index]
confidence = float(predictions[0][predicted_index])

# -----------------------------
# Show All Images (Including Augmented)
# -----------------------------
plt.figure(figsize=(15, 6))

# Row 1: Original + Processing
plt.subplot(2, 3, 1)
plt.imshow(original_img)
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 3, 2)
plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale")
plt.axis("off")

plt.subplot(2, 3, 3)
plt.imshow(enhanced_array, cmap='gray')
plt.title("Enhanced (CLAHE)")
plt.axis("off")

# Row 2: Augmented Samples
for i in range(3):
    plt.subplot(2, 3, i + 4)
    plt.imshow(augmented_samples[i], cmap='gray')
    plt.title(f"Augmented #{i+1}")
    plt.axis("off")

plt.suptitle(
    f"True: {true_label} | Predicted: {predicted_label} ({confidence:.2f})",
    fontsize=12
)
plt.show()

# -----------------------------
# Debug Output
# -----------------------------
print("Enhanced Input Shape:", img_array.shape)
print("Min, Max pixel values:", img_array.min(), img_array.max())
print("Predictions:", predictions)
```
